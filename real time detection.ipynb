```python
import os  
import cv2  
from mss import mss
from PIL import Image
from PIL import ImageGrab
import numpy as np  
import tensorflow as tf
import keras
from keras.models import model_from_json  
from tensorflow.keras.utils import img_to_array
```

```python
#load model  
#change directory to correctly import model
model = model_from_json(open(r"vgg16 network\vgg16_v2.json", "r").read())  


#load weights  
#change directory to correctly import weights
model.load_weights(r'vgg16 network\vgg16_fine_tuned_v2.h5') 
```

```python
#import haar cascade
#change directory to correctly import haar features
face_haar_cascade = cv2.CascadeClassifier(r'C:\Users\abhin\OneDrive\Desktop\The folder of folders\vs\dissertation\haarcascade_frontalface_default.xml')  
  
#video capture 
cap=cv2.VideoCapture(0)  
  
#while loop to display camera feed for real time detection  
while True:  
    ret,test_img=cap.read()# captures frame and returns boolean value and captured image  
    if not ret:  
        continue  
    gray_img= cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)  
  
    faces_detected = face_haar_cascade.detectMultiScale(gray_img, 1.32, 5)  
  
  
    for (x,y,w,h) in faces_detected:  
        cv2.rectangle(test_img,(x,y),(x+w,y+h),(255,0,0),thickness=7)  
        roi_gray=gray_img[y:y+w,x:x+h]#cropping region of interest i.e. face area from  image  
        roi_gray=cv2.resize(roi_gray,(75,75))  
        img_pixels = img_to_array(roi_gray)  
        img_pixels = np.expand_dims(img_pixels, axis = 0)
  
        predictions = model.predict(img_pixels)  
  
        #find max indexed array  
        max_index = np.argmax(predictions[0])  
  
        emotions = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')  
        predicted_emotion = emotions[max_index]  
  
        cv2.putText(test_img, predicted_emotion, (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)  
  
    resized_img = cv2.resize(test_img, (1000, 700))  
    cv2.imshow('Facial emotion analysis ',resized_img)  
  
  
  
    if cv2.waitKey(10) == ord('q'):#wait until 'q' key is pressed  
        break  

cap.release() 
cv2.destroyAllWindows  
```
